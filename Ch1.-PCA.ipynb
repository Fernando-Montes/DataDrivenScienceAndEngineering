```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

plt.rcParams["figure.figsize"] = [6, 6]
plt.rcParams["figure.autolayout"] = True
```

```python
# Generate artificial data
n = 10000
theta = np.pi/3
mu = [2 , 1]
sigma = [2, 0.5]
R = [[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]]
X = np.asarray( [ mu + np.dot(R, sigma*np.random.normal(size=2)) for i in range(n) ] )
plt.scatter(X[:,0], X[:,1], s = 5)
plt.xlim(-6,8)
plt.ylim(-6,8)
plt.show()
```

```python
# Compute PCA
X = mu + StandardScaler(with_std = False).fit_transform(X)   # PCA subtracts the mean so need to add it again
pca = PCA(n_components = 2).fit(X)
print(R)
print(pca.components_)
print(pca.explained_variance_)
print(pca.singular_values_)

v1 = sigma[0]*pca.components_[0, :] 
v2 = sigma[1]*pca.components_[1, :] 

plt.scatter(X[:,0], X[:,1], s = 5)
plt.plot([mu[0], mu[0] + v1[0]], [mu[1], mu[1] + v1[1]], color = 'black')
plt.plot([mu[0], mu[0] + v2[0]], [mu[1], mu[1] + v2[1]], color = 'black')
plt.xlim(-6,8)
plt.ylim(-6,8)
plt.show()
```

```python
# Visualize PCA
Xpca = pca.transform(X)
plt.scatter(Xpca[:,0], Xpca[:,1], s = 5)
plt.show()
```
